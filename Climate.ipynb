```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
%matplotlib notebook

Dir = 'C:/Illisarvik/ClimateData'

for i,filename in enumerate(os.listdir(Dir)):
    F = pd.read_csv(Dir+'/'+filename,header=22)
    F = F.set_index(pd.DatetimeIndex(F['Date/Time'],drop=True)).drop(['Date/Time'],axis=1)
#     F=F
    if i == 0:
        Data = F.copy()
    else:
        Data = Data.append(F)
Data['DOY']=Data.index.dayofyear
Daily=Data.groupby('DOY').mean()
Monthly = Data.groupby('Month').mean()
MonthlySum = Data.groupby('Month').sum()
Monthly['Total Precip (mm)']=MonthlySum['Total Precip (mm)']#/10
# print(Monthly.head())
fig,ax = plt.subplots(2,1,figsize=(6.5,7))
plt.figure()
plt.plot(Data['Total Precip (mm)'])
plt.grid()


print(Monthly['Mean Temp (Â°C)'].mean())
print(Monthly['Total Precip (mm)'].sum())
# Total Precip (mm)
```
